## University Student Registration System
### Add Students
### View GPA
### Assign Academic Violations

```python



fname = ["Carmen","Sydney","Neil","Marcus"]
lname = ["Berzatto","Adamu","Fak","Brooks"]
major = ["Marketing","Food History","Video Game Repair","Baking"]
gpa = [2.5,3.9,2.0,3.3]
violations = ["Plagiarism","None","Copying Homework","None"]
infractions = ["Copying Homework", "Cheating on Exam", "Plagiarism", "Cheating on Quiz", "Unruly Behavior", "None"]

x = 0
print("Student Information System")
print("Select an option:")

while x<5:
    print("1.Student Information Review")
    print("2.Student Information Entry")
    print("3.GPA Metrics")
    print("4.Academic Violations")
    print("5.Quit Program")
    x = int(input("Select an Option\n"))
    if x >= 5:
        break

    ##This section prints the summary of all students in the system
    
    if x == 1:
        length = len(fname)
        i=0
        while i<length:
            print(fname[i],lname[i],", Major:",major[i],", GPA",gpa[i])
            i = i+1

    ##This section allows user to input a new student into the system, with GPA and Major also        
    
    elif x == 2: 
        addfname = str(input("Input Student's First Name"))
        addlname = str(input("Input Student's Last Name"))
        addmajor = str(input("Input Student's Major"))
        addgpa = float(input("Input Student's GPA"))
        fname.append(addfname)
        lname.append(addlname)
        major.append(addmajor)
        gpa.append(addgpa)
        violations.append("None")
        print(addfname,addlname,"added to system")

    ##This returns summary GPA stats to the user
    
    elif x == 3: 
        lengpa = len(gpa)
        sumgpa = 0
        maxgpa = 0
        mingpa = 4
        j=0
        while j < lengpa:
            if gpa[j] > maxgpa:
                maxgpa = gpa[j]
            if gpa[j] < mingpa:
                mingpa = gpa[j]
            sumgpa = sumgpa + gpa[j]
            j = j + 1
        avggpa = sumgpa / lengpa
        print("Average GPA for Students is", avggpa)
        print("Max GPA is ", maxgpa)
        print("Min GPA is ", mingpa)

    ##This section allows user to add an infraction to a student in the system, then view all student infractions
    
    else:
        vioaddf = str(input("Input Student's First Name to Add Infraction"))
        vioaddl = str(input("Input Student's Last Name to Add Infraction"))
        print("Infractions",infractions)
        infadd = str(input("Input Infraction"))
        k=0
        while k < len(violations):
            if vioaddf == fname[k] and vioaddl == lname[k]:
                violations[k] = infadd
                print(infadd,"Infraction added to record of",vioaddf,vioaddl)
            k = k+1
        length = len(violations)
        y=0
        while y<length:
            print(fname[y],lname[y],", Infraction:",violations[y])
            y = y+1
        
print("Have a Nice Day") 
    
```
