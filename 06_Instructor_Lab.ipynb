## Magic Crystal Ball- Simple $30 Fee to Answer Life's Most Mysterious Questions

```python
import random as rand

outputs = [
    "It is certain", 
    "It is decidedly so", 
    "That question requires greater monetary compensation", 
    "Reply is hazy; try again with more money", 
    "Concentrate and ask again", 
    "My reply is no", 
    "Outlook is not so good", 
    "Very Doubtful", 
    "You would like to know that, wouldn't you?", 
    "It's never that simple, to simply answer such a question!",
    "Why don't you go find the answer yourself?"]

print("Welcome to Winston's the Wizard's Wonderful Worldly Predictor")
print("Would you like to:")
print("1. Ask me a question, costing you $30 per question?")
print("2.Review your Questions and Responses")
print("3.Quit services and Pay?")
print("Select 1. 2. or 3.")

payment = 0.0
questions = []
answers = []

i=0
while i<1:
    sel = int(input("Ask a Question\n"))
    print(sel)
    if sel == 1:
       randA = rand.randrange(0,len(outputs))
       payment = payment + 30
       curque = str(input("Ask your Question!\n"))
       curans = outputs[randA]
       answers.append(curans)
       questions.append(curque)
       print(curans)
       print("1.Ask Another Question")
       print("2.Review your Questions and Responses")
       print("3.Quit services and Pay?")
    elif sel == 2:
        j=0
        while j<len(questions):
            print("Question:",questions[j])
            print("Answer:",answers[j])
            j=j+1
        print("1.Ask Another Question")
        print("2.Review your Questions and Responses")
        print("3.Quit services and Pay?")
    elif sel == 3:
        print("You must now pay me or suffer the consequences!")
        print("You asked ",len(questions), "questions at $30 each")
        print("Remit payment of $", payment,"Immediately")
        i=i+1
    else:
        print("That was not one of your options! Try again")
```
